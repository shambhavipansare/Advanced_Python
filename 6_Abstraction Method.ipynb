```python
from abc import ABC, abstractmethod
```

```python
class car(ABC):
    @abstractmethod
    def airbag(self):         # compulsory function
        pass
```

```python
class nexon(car):
    def airbag(self):
        return "Airbag loaded in tata nexon"
    def sunroof(self):
        return "Sunroof is opening"
```

```python
obj1 = nexon()
```

```python
obj1.sunroof()
```

```python
obj1.airbag()
```

```python
class fortuner(car):
    def key_start(self):
        return 'Car is starting'
```

```python
f = fortuner()
```

```python
class phone(ABC):
    @abstractmethod
    def calling(self):
        pass
```

```python
class samsung(phone):
    def calling(self):
        return 'Samsung is going to call'
    def camera(self):
        return 'Camera is opening'
```

```python
p1 = samsung()
```

```python
p1.calling()
```

```python
p1.camera()
```

```python
class iphone(phone):
    def play_music(self):
        return 'Music is playing'
```

```python
i1 = iphone()
```

```python
# base class

class car(ABC):
    @abstractmethod
    def start(self):
        pass
```

```python
class suzuki(car):
    def start(self):
        return 'Suzuki is going to start car'
    def ac(self):
        return 'AC is started'      
```

```python
swift = suzuki()
```

```python
swift.start() 
```

```python
class tata(car):
    def start(self):
        return 'Tata is going to start your car'
    def autopark(self):
        return 'Car is going to park'
```

```python
altoz = tata()
```

```python
altoz.start()
```

```python
altoz.autopark()
```

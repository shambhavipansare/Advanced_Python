```python
class cal():
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def add(self):
        print(self.a + self.b)
    def mul(self):
        print(self.a * self.b)
    def div(self, c):
        print((self.a / self.b) + c)
```

```python
ob1 = cal(10, 5)
```

```python
ob1.a
```

```python
ob1.b
```

```python
ob1.add()
```

```python
ob1.mul()
```

# Incapsulation

```python
class car():
    def __init__(self, fuel, color, engine):        # constructor
        self.fuel = fuel                            # public
        self._color = color                         # protected
        self.__engine = engine                      # private
```

```python
c1 = car('Petrol', 'Black', 1000)
```

```python
c1.fuel
```

```python
c1._color
```

```python
c1._car__engine
```

# Inheritance

```python
class parent():
    def singing(self):
        print('I am from parent class with singing skills')
```

```python
class child(parent):
    def cricket(self):
        print('I am from child class with cricket skills')
```

```python
ob = child()
```

```python
ob.cricket()
```

```python
ob.singing()
```

```python
p1 = parent()
```

```python
p1.singing()
```

```python
class next_child(child):
    def dancing(self):
        print('I am from next_child class with dancing skills')
```

```python
nc = next_child()
```

```python
nc.singing()
```

```python
nc.dancing()
```

```python
nc.cricket()
```

# Polyorphism

```python
class animal():
    def sound(self):
        print('Animal will make sound')
```

```python
class dog():
    def sound(self):
        print('Bhau - Bhau')
```

```python
class cat():
    def sound(self):
        print('Meow - Meow')
```

```python
for i in [animal(), dog(), cat()]:
    print(i.sound())
```

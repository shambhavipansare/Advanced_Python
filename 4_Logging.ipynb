```python
import logging

logging.basicConfig(filename = 'project.log', level = logging.debug, format = '%(levelname)s - %(asctime)s - %(message)s')

def add(x, y):
    try:
        logging.info('Project 1 started')
        logging.info(f'adding {x} and {y}')
        result = x + y
        logging.info(f'Result is {result}')
        logging.info('Project 1 completed')
    except Exception as e:
        logging.error(e)
```

```python
add(3, 5)
```

```python
add('Hello', 5)
```
